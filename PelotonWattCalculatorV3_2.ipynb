The purpose of this project was to create a model similar to what Peloton uses to calculate power in Watts from a Gen2 Peloton utilizing Tensorflow's framework to code an artificial neural network.

The magnetic resistance adjustment on the Gen2 is not standard and, thus, the method of power calculation is proprietary and remains unknown/unclear based on various internet searches. 

Additional hurdles included an absence of any Peloton API to obtain direct data from the bike. My wife did not want me to hack the bike and possibly brick it (since it was fairly expensive), therefore a work around using optical character recognition was used. 

Overall strategy was to extract the necessary data from video, process the data, and finally train a neural network to model the data and approximate the algorithm

Data collection method:

I set up a camera to record the screen and varied my Cadence from 60-100 through every 2 values of resistance:

60-100 at Resistance 24
60-100 at Resistance 26
60-100 at Resistance 28
60-100 at Resistance 30

Towards the end I attempted to hold cadence around 60-70 and then slowly increased resistance from 20-50 in order to try and keep a "constant" cadence sample for a relationship between Power and Resistance only

```python
#create empty arrays to store the upcoming cropped images

cadence_array = []
power_array = []
resistance_array = []
```

```python
# video is 30 FPS therefore will need to keep every 1 image every second therefore %30

capture_file = cv2.VideoCapture('/content/drive/MyDrive/Colab Notebooks/pelotonCapMasked.m4v')

ret = True
counter = 0

while ret == True:
    
    print(f'# {counter}')
    #read frame, ret True = not at end of file
    ret,frame = capture_file.read()
    
    #save every second, every 30 files
    if (counter%30 == 0):
        # append crop to cadence array
        cadence_array.append(frame[190:300,80:200])

        # append crop to power array
        power_array.append(frame[150:300,230:390])

        # appnd crop to resistance array
        resistance_array.append(frame[150:300,410:550])
    counter += 1
    
capture_file.release()
```

```python
plt.imshow(resistance_array[200])
```

```python
# instantiate OCR reader object to look through the arrays
ocrReader = easyocr.Reader(['en'])
```

```python

#create the arrays to store OCR recognized data
cadenceOCRd = []
powerOCRd = []
resistanceOCRd = []

counter = 0
cadence_array_length = len(cadence_array)

#loop through each crop array and read the number and store to new OCR array we just made
#anything it had trouble reading assign a (-1) to later filter out

for i in range(len(cadence_array)):
    
    print(f'{counter} of {cadence_array_length}')
    counter = counter + 1

    #cadence
    number = ocrReader.readtext(cadence_array[i], detail=0)
    #make int and append
    try:
        cadenceOCRd.append(number[0])
    except:
        cadenceOCRd.append(-1)
    
    #power
    number = ocrReader.readtext(power_array[i], detail=0)
    try:
        powerOCRd.append(number[0])
    except:
        powerOCRd.append(-1)
    
    #resistance
    number = ocrReader.readtext(resistance_array[i], detail=0)
    try:
        resistanceOCRd.append(number[0])
    except:
        resistanceOCRd.append(-1)
```

```python

# Convert all INT to FLOATS 

powerOCRdFloat = []

for i in powerOCRd:
    try:
        i = float(i)
    except:
        i = -1
    powerOCRdFloat.append(i)

cadenceOCRdFloat = []

for i in cadenceOCRd:
    try:
        i = float(i)
    except:
        i = -1
    cadenceOCRdFloat.append(i)

resistanceOCRdFloat = []

for i in resistanceOCRd:
    try:
        i = float(i)
    except:
        i = -1
    resistanceOCRdFloat.append(i)
```

```python
# check to make sure it's float... YES!!!

type(resistanceOCRdFloat[0])
```

```python
# create Dictionary with data

pelotonDict = {'Cadence': cadenceOCRdFloat, 'Power': powerOCRdFloat, 'Resistance': resistanceOCRdFloat}
```

```python
# convert to a Pandas DataFrame

pelotonDF = pd.DataFrame(pelotonDict)
```

Power is the targeted Feature


```python
# prep data
X = pelotonDF.drop('Power', axis=1).values
y = pelotonDF['Power'].values
```

```python
# matrix of cadence and resistance

X
```

```python
# array of output Power matching up with above X matrix

y
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
# split data up into training set and testing set

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.33, random_state=42)
```

```python
# verified X split to total 393 data points previously
len(X_train)+len(X_test)
```

```python
# Verify y as well, also 393 so we haven't lost anything
len(y_train) + len(y_test)
```

```python
# Scale the data to between 0-1
scaler = MinMaxScaler(feature_range=(0,1))
```

```python
# fitting and scaling data to the X_training set

X_train_scaled = scaler.fit_transform(X_train)
```

```python
# check scaled X... looks good

X_train_scaled
```

```python
# Transform the test data using the X_train scaler

X_test_scaled = scaler.transform(X_test)
```

```python
# scaled X_test data also looks good

X_test_scaled
```

```python
# import Tensorflow libraries to create the ANN
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dense
from tensorflow.keras.optimizers import Adam

```

```python
# check number of data points in training set split from earlier

len(X_train_scaled)
```

```python
# check data shape to input into ANN

X_train_scaled.shape
```

```python
# build Tensorflow ANN model using 4 layers neurons
model = Sequential([
    Dense(units=32, input_shape=(2,), activation='relu'),
    Dense(units=32, activation='relu'),
    Dense(units=32, activation='relu'),
    Dense(units=1)
])
```

```python
# output 1 neuron

model.summary()
```

```python
# compile model using adam optimizer and MSE due to continuous data
# 750 might be overkill but will tweak later

model.compile(optimizer='adam', loss='mean_squared_error', metrics=['accuracy'])
```

```python
# train the model, mechine is LEARNING!!!
epoch_history = model.fit(x=X_train_scaled, y=y_train, batch_size=64, validation_split=0.1, epochs=750, verbose=2)
```

```python
print(epoch_history.history.keys())
```

```python
plt.plot(epoch_history.history['loss'])
plt.plot(epoch_history.history['val_loss'])
plt.title('Model Loss Progress During Training')
plt.xlabel('Epoch Number')
plt.ylabel('Training and Validation Loss')
plt.legend(['Training Loss','Validation Loss'])

# As time progresses the machine makes less errors as it learns and therefore the plot drops to 0
```

```python
# now feed in the testing batch that the ANN has never seen and let the neural network try to figure 
# out the test data and make predictions

predictions = model.predict(x=X_test_scaled, batch_size=10,verbose=2)
```

```python
plt.figure(figsize=(20,10))

plt.plot(y_test, color='blue')
plt.plot(predictions, color='red')
plt.title('Neural Network Predictions vs True Values')
plt.xlabel('Epoch Number')
plt.ylabel('Power in Watts')
plt.legend(['Testing Data', 'Neural Network Predictions'])

# for the most part they almost match up! Artificial Neural Network is doing pretty well!!!
```
